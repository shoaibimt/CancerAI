```python
target = new_client.target
target_query_bcrabl = target.search('BCR-ABL')
bcrabl_data = pd.DataFrame.from_dict(target_query_bcrabl)
bcrabl_data
```

```python
#CHEMBL2096618
selected_target = bcrabl_data.target_chembl_id[2]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```
