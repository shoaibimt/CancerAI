```python
import padelpy
from padelpy import padeldescriptor
```

```python
for i in fp:
    path = "/content/drive/MyDrive/inhibitor_prediction/bcr_abl/chembl_data_retrieve/mol_fingerprints/Results"
    fingerprint = i
    fingerprint_output_file_path = ''.join([path,fingerprint])
    fingerprint_output_file = ''.join([fingerprint_output_file_path,'.csv'])
    fingerprint_descriptortypes = fp[fingerprint]

    padeldescriptor(mol_dir='/content/drive/MyDrive/inhibitor_prediction/bcr_abl/chembl_data_retrieve/molecule.smi',
                    d_file=fingerprint_output_file,
                    descriptortypes= fingerprint_descriptortypes,
                    detectaromaticity=True,
                    standardizenitro=True,
                    standardizetautomers=True,
                    threads=2,
                    removesalt=True,
                    log=True,
                    fingerprints=True)
```

```python
# Create a list with shorten names of xml files
FP_list = ['AtomPairs2DCount',
 'AtomPairs2D',
 'EState',
 'CDKextended',
 'CDK',
 'CDKgraphonly',
 'MACCS',
 'PubChem',
 'SubstructureCount',
 'Substructure']
```

```python
# Load original molecular descriptor files and store them in a dictionary
descriptors = {}

for i in FP_list:
    path = "/content/drive/MyDrive/inhibitor_prediction/bcr_abl/chembl_data_retrieve/mol_fingerprints/Results/"
    fingerprint_file_path = ''.join([path,i])

    descriptor_name_csv = ''.join([fingerprint_file_path,'.csv'])

    print(descriptor_name_csv)

    descriptors[str(i)] = pd.read_csv(descriptor_name_csv)

    descriptors[str(i)] = descriptors[str(i)].sort_values("Name").reset_index(drop=True)
```
